```python
import pandas as pd
from selenium import webdriver
import time
```

```python
namelist = []
teamlist = []
scorelist = []
setcount = []
errlist = []

attlist = []
atslist = []
#atblist = []
atelist = []
svtlist = []
svslist = []
sttlist = []
stslist = []

bltlist = []
blslist = []
rstlist = []
rsclist = []
rsflist = []
dgtlist = []
dgslist = []
```

```python
browser = webdriver.Chrome('C:/Users/ADMIN/Desktop/chromedriver.exe')  # chromedriver 다운받고, 다운 받은 경로 써주어야함

# 크롤링 할 포지션, 페이지 수 입력
position = input("포지션: ")
page = int(input("페이지 수: "))

pgcount=1
# 페이지 수만큼 반복문 돌리기
while(pgcount<=page):
    
    # 입력받은 포지션, 현재 크롤링 할 페이지 url에 적용
    search_url = "http://www.welcomebanktoprankings.co.kr/volleyball/index.php?p=chart&season=divKOVOS&searchType=divKOVOS&searchDate=SEASON2022&position="+position+"&tpage=2&page="+str(pgcount)
    pgcount+=1

    browser.get(search_url)
    browser.implicitly_wait(2)

    
    # 크롤링 할 데이터 테이블,태그 찾기 - 선수 정보 (이름, 소속 구단, 참여한 세트 수 등)
    ptable = browser.find_element_by_class_name('table-div')
    ptbody = ptable.find_element_by_tag_name("tbody")
    prows = ptbody.find_elements_by_tag_name("tr")

    # 반복문 돌려서 행마다 필요한 데이터 모으기
    for index, value in enumerate(prows):
        pbody = value.find_elements_by_tag_name("td")[1:]
        
        # 데이터 변수에 저장 -> 리스트에 넣기 | 이중으로 작업했지만 바로 리스트에 넣어도 됨
        name = pbody[0].text
        team = pbody[1].text
        score = float(pbody[2].text)
        stcount = int(pbody[4].text)
        error = int(pbody[6].text)
        
        namelist.append(pbody[0].text)
        teamlist.append(team)
        scorelist.append(score)
        setcount.append(stcount)
        errlist.append(error)
        
        
    # 크롤링 할 데이터 테이블,태그 찾기 - 경기 데이터 (공격, 서브, 세트 등 공격 지표로 사용될 수 있음)
    atable = browser.find_elements_by_class_name('type01')[1]
    atbody = atable.find_elements_by_tag_name("tbody")[1]
    arows = atbody.find_elements_by_tag_name("tr")
    
    for index, value in enumerate(arows):
        abody = value.find_elements_by_tag_name("td")
        
        attry = int(abody[0].text)
        atsucc = int(abody[1].text)
        #atblock = int(abody[2].text)    # 블로킹에 막힌 공격 수는 불필요해서 제외함
        aterr = int(abody[3].text)
        svtry = int(abody[4].text)
        svsucc = int(abody[5].text)
        sttry = int(abody[7].text)
        stsucc = int(abody[8].text)
        
        attlist.append(attry)
        atslist.append(atsucc)
        #atblist.append(atblock)
        atelist.append(aterr)
        svtlist.append(svtry)
        svslist.append(svsucc)
        sttlist.append(sttry)
        stslist.append(stsucc)
    
    
    # 경기 데이터 - 다음 테이블로 이동
    browser.find_elements_by_class_name('table-btn')[1].click()
    time.sleep(1)
    
    
    # 크롤링 할 데이터 테이블,태그 찾기 - 경기 데이터 (블로킹, 리시브, 디그 등 수비 지표로 사용될 수 있음)
    dtable = browser.find_elements_by_class_name('type01')[2]
    dtbody = dtable.find_elements_by_tag_name("tbody")[1]
    drows = dtbody.find_elements_by_tag_name("tr")
    
    for index, value in enumerate(drows):
        dbody = value.find_elements_by_tag_name("td")
        bltry = int(dbody[0].text)
        blsucc = int(dbody[1].text)
        rstry = int(dbody[3].text)
        rscorr = int(dbody[4].text)
        rsfail = int(dbody[5].text)
        dgtry = int(dbody[6].text)
        dgsucc = int(dbody[7].text)
        
        bltlist.append(bltry)
        blslist.append(blsucc)
        rstlist.append(rstry)
        rsclist.append(rscorr)
        rsflist.append(rsfail)
        dgtlist.append(dgtry)
        dgslist.append(dgsucc)


# 크롤링 끝
browser.close()
```

```python
# DataFrame - Libero

data = {"name":namelist,"team":teamlist,"pos":"Li","score":scorelist,"error":errlist,"setcount":setcount,
        "at_try":attlist,"at_succ":atslist,"at_err":atelist,"sv_succ":svslist,"set_succ":stslist,"bl_succ":blslist,
        "rs_try":rstlist,"rs_corr":rsclist,"rs_fail":rsflist,"dg_succ":dgslist}

libero = pd.DataFrame(data)
libero.index += 1
libero.head(5)

#libero.to_csv('C:\\Users\\ADMIN\\data\\FeVolleyball\\libero.csv',encoding='utf-8-sig')
```
